```python
import numpy as np
import matplotlib.pyplot as plt
import shapely.geometry as shpg
```

```python
# 1. import cortical ext + int slices
ext_slice = np.load(r'/home/simoneponcioni/Documents/01_PHD/03_Methods/Meshing/Meshing/99_testing_prototyping/testing_transfinite_2D/C0002233_int_CORT_ext_8_point_insertion.npy')
int_slice = np.load(r'/home/simoneponcioni/Documents/01_PHD/03_Methods/Meshing/Meshing/99_testing_prototyping/testing_transfinite_2D/C0002233_int_CORT_int_8_point_insertion.npy')
```

```python
extpol = shpg.polygon.Polygon(ext_slice)
intpol = shpg.polygon.Polygon(int_slice)
cortex = extpol.difference(intpol)
cortex_centroid = cortex.centroid.coords.xy[0][0], cortex.centroid.coords.xy[1][0]
print(type(cortex_centroid))
```

```python
ax = plt.subplots(1, figsize=(10, 10))[1]
plt.plot(ext_slice[:, 0], ext_slice[:, 1], c='r')
plt.plot(int_slice[:, 0], int_slice[:, 1], c='b')
plt.scatter(cortex_centroid[0], cortex_centroid[1], c='k', marker='x', label='cortex centroid')
plt.legend()
ax.set_title('CORTICAL EXT + INT SLICES')
ax.set_aspect('equal')
```
